## Задание №1

Реализовать и сравнить работу различных алгоритмов поиска в списке и в бинарном дереве.


```python
import random, time

# 1. Генерация случайных данных
data = sorted([random.randint(1, 1000) for _ in range(100)])
print('Исходные данные (отсортированы):')
print(data)
print()

# 2. Бинарный поиск
def binary_search(arr, x):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# 3. Фибоначчиев поиск
def fibonacci_search(arr, x):
    n = len(arr)
    fibMMm2, fibMMm1 = 0, 1
    fibM = fibMMm2 + fibMMm1
    while fibM < n:
        fibMMm2, fibMMm1 = fibMMm1, fibM
        fibM = fibMMm2 + fibMMm1

    offset = -1
    while fibM > 1:
        i = min(offset + fibMMm2, n - 1)
        if arr[i] < x:
            fibM, fibMMm1 = fibMMm1, fibMMm2
            fibMMm2 = fibM - fibMMm1
            offset = i
        elif arr[i] > x:
            fibM, fibMMm1 = fibMMm2, fibMMm1 - fibMMm2
            fibMMm2 = fibM - fibMMm1
        else:
            return i
    if fibMMm1 and offset + 1 < n and arr[offset + 1] == x:
        return offset + 1
    return -1

# 4. Интерполяционный поиск
def interpolation_search(arr, x):
    low, high = 0, len(arr) - 1
    while low <= high and arr[low] <= x <= arr[high]:
        if low == high:
            return low if arr[low] == x else -1
        pos = low + ((x - arr[low]) * (high - low) // (arr[high] - arr[low]))
        if pos >= len(arr):
            return -1
        if arr[pos] == x:
            return pos
        if arr[pos] < x:
            low = pos + 1
        else:
            high = pos - 1
    return -1

# 5. Бинарное дерево поиска
class Node:
    def __init__(self, key):
        self.left = self.right = None
        self.val = key

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert(key, self.root)

    def _insert(self, key, node):
        if key < node.val:
            if node.left is None:
                node.left = Node(key)
            else:
                self._insert(key, node.left)
        elif key > node.val:
            if node.right is None:
                node.right = Node(key)
            else:
                self._insert(key, node.right)

    def search(self, key):
        return self._search(key, self.root)

    def _search(self, key, node):
        if node is None or node.val == key:
            return node
        elif key < node.val:
            return self._search(key, node.left)
        else:
            return self._search(key, node.right)

    def delete(self, key):
        self.root = self._delete(key, self.root)

    def _delete(self, key, node):
        if node is None:
            return node
        elif key < node.val:
            node.left = self._delete(key, node.left)
        elif key > node.val:
            node.right = self._delete(key, node.right)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            temp = self._find_min(node.right)
            node.val = temp.val
            node.right = self._delete(temp.val, node.right)
        return node

    def _find_min(self, node):
        while node.left:
            node = node.left
        return node

# Создание дерева и вставка элементов
bst = BinarySearchTree()
for num in data:
    bst.insert(num)

# 6. Сравнение всех методов поиска
search_key = random.choice(data)
print(f'\nПоиск элемента: {search_key}\n')

# Бинарный поиск
start = time.time()
res = binary_search(data, search_key)
end = time.time()
print(f'Бинарный поиск: найден индекс {res}, время {end - start:.6f} сек.')

# Фибоначчиев поиск
start = time.time()
res = fibonacci_search(data, search_key)
end = time.time()
print(f'Фибоначчиев поиск: найден индекс {res}, время {end - start:.6f} сек.')

# Интерполяционный поиск
start = time.time()
res = interpolation_search(data, search_key)
end = time.time()
print(f'Интерполяционный поиск: найден индекс {res}, время {end - start:.6f} сек.')

# Поиск в бинарном дереве
start = time.time()
res = bst.search(search_key)
end = time.time()
found = res is not None
print(f'Поиск в бинарном дереве: найден = {found}, время {end - start:.6f} сек.')

# Обычный поиск через in
start = time.time()
found = search_key in data
end = time.time()
print(f'Поиск через "in": найден = {found}, время {end - start:.6f} сек.')

# Удаление элемента из дерева
bst.delete(search_key)
print(f'\nУдаление элемента {search_key} из бинарного дерева выполнено.')
res = bst.search(search_key)
print('Проверка после удаления:', 'Не найден' if res is None else 'Найден')
```

# Задание №2


Реализовать три различных метода хеширования для поиска, вставки и удаления элементов в хеш-таблице, а также провести их сравнение по нескольким критериям, чтобы оценить эффективность каждого метода.

```python
import time
import random

TABLE_SIZE = 50
NUM_KEYS = 30
random.seed(42)

keys = random.sample(range(1000), NUM_KEYS)
values = [f"Val{key}" for key in keys]
test_key = keys[15]

# -----------------------------
# Подсчёт коллизий
collisions = {
    "simple": 0,
    "random": 0,
    "chain": 0
}

# -----------------------------
# Простое хеширование (линейное пробирование)
simple_table = [None] * TABLE_SIZE

def simple_hash(key):
    return key % TABLE_SIZE

def simple_insert(key, value):
    index = simple_hash(key)
    start_index = index
    while simple_table[index] is not None:
        if simple_table[index][0] == key:
            simple_table[index] = (key, value)
            return
        index = (index + 1) % TABLE_SIZE
        collisions["simple"] += 1
        if index == start_index:
            return
    simple_table[index] = (key, value)

def simple_search(key):
    index = simple_hash(key)
    start_index = index
    while simple_table[index] is not None:
        if simple_table[index][0] == key:
            return simple_table[index][1]
        index = (index + 1) % TABLE_SIZE
        if index == start_index:
            break
    return None

for k, v in zip(keys, values):
    simple_insert(k, v)

start = time.perf_counter()
simple_result = simple_search(test_key)
time_simple = time.perf_counter() - start

# -----------------------------
# Рандомизированное пробирование
random_table = [None] * TABLE_SIZE

def random_hash(key):
    return (key * 17 + 13) % TABLE_SIZE

def random_insert(key, value):
    index = random_hash(key)
    i = 1
    while random_table[index] is not None:
        if random_table[index][0] == key:
            random_table[index] = (key, value)
            return
        index = (index + i * 3) % TABLE_SIZE
        i += 1
        collisions["random"] += 1
    random_table[index] = (key, value)

def random_search(key):
    index = random_hash(key)
    i = 1
    while random_table[index] is not None:
        if random_table[index][0] == key:
            return random_table[index][1]
        index = (index + i * 3) % TABLE_SIZE
        i += 1
    return None

for k, v in zip(keys, values):
    random_insert(k, v)

start = time.perf_counter()
random_result = random_search(test_key)
time_random = time.perf_counter() - start

# -----------------------------
# Хеширование с цепочками
chain_table = [[] for _ in range(TABLE_SIZE)]

def chain_insert(key, value):
    index = key % TABLE_SIZE
    for i, (k, v) in enumerate(chain_table[index]):
        if k == key:
            chain_table[index][i] = (key, value)
            return
    if chain_table[index]:  # если список уже не пустой — это коллизия
        collisions["chain"] += 1
    chain_table[index].append((key, value))

def chain_search(key):
    index = key % TABLE_SIZE
    for k, v in chain_table[index]:
        if k == key:
            return v
    return None

for k, v in zip(keys, values):
    chain_insert(k, v)

start = time.perf_counter()
chain_result = chain_search(test_key)
time_chain = time.perf_counter() - start

# -----------------------------
# Оценка по критериям
def time_grade(t):
    if t < 0.000005:
        return "A"
    elif t < 0.00001:
        return "B"
    elif t < 0.00005:
        return "C"
    else:
        return "D"

def collision_grade(c):
    if c <= 3:
        return "A"
    elif c <= 10:
        return "B"
    elif c <= 20:
        return "C"
    else:
        return "D"

def complexity_grade(method):
    if method == "simple":
        return "A"  # простое
    elif method == "random":
        return "C"  # посложнее
    elif method == "chain":
        return "B"  # среднее

# -----------------------------
# Вывод таблицы

print("\nОценка эффективности методов хеширования:\n")
print(f"{'Метод':<35} {'Время':<10} {'Коллизии':<10} {'Простота':<10} {'Итог'}")
print("-" * 70)

for method in ["simple", "random", "chain"]:
    t = {"simple": time_simple, "random": time_random, "chain": time_chain}[method]
    time_g = time_grade(t)
    coll_g = collision_grade(collisions[method])
    comp_g = complexity_grade(method)

    # Итог: средняя оценка (грубо: B = 2, C = 3 и т.д.)
    grades = [time_g, coll_g, comp_g]
    score_map = {"A": 1, "B": 2, "C": 3, "D": 4}
    avg_score = round(sum(score_map[g] for g in grades) / 3)
    final_grade = [k for k, v in score_map.items() if v == avg_score][0]

    method_name = {
        "simple": "Простое хеширование",
        "random": "Рандомизированное хеширование",
        "chain": "Хеширование с цепочками"
    }[method]

    print(f"{method_name:<35} {time_g:<10} {coll_g:<10} {comp_g:<10} {final_grade}")

```

## Задание №3

Найти хотя бы одно (или все возможные) правильное расположение ферзей, где они не угрожают друг другу. Важно, что это задание решается с помощью методов поиска, таких как поиск с возвратом (backtracking).

```python
# Функция для проверки, не угрожает ли ферзь на позиции (row, col) другим ферзям
def is_safe(board, row, col):
    # Проверка столбца
    for i in range(row):
        if board[i] == col or abs(board[i] - col) == abs(i - row):
            return False
    return True

# Рекурсивная функция для размещения ферзей
def solve_nqueens(board, row):
    # Если все ферзи размещены, выводим решение
    if row == 8:
        print_board(board)
        return True
    
    # Пробуем разместить ферзя в каждом столбце текущей строки
    for col in range(8):
        if is_safe(board, row, col):
            board[row] = col
            if solve_nqueens(board, row + 1):
                return True
            board[row] = -1  # Если не удалось разместить, откатываем изменения
    return False

# Функция для вывода доски
def print_board(board):
    for i in range(8):
        row = ['.' for _ in range(8)]  # Инициализация пустой строки
        row[board[i]] = 'Ф'  # Размещение ферзя
        print(" ".join(row))
    print()

# Главная функция для решения задачи
def solve():
    board = [-1] * 8  # Создаем пустую доску, -1 означает отсутствие ферзя в строке
    if not solve_nqueens(board, 0):
        print("Решение не найдено.")
    else:
        print("Решение найдено.")

# Запуск программы
solve()
```
